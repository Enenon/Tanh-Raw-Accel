```python
# no cmd digite: pip install matplotlib; pip install ipywidgets

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
from ipywidgets import FloatSlider
```

Criação dos vetores e constantes iniciais

```python
x = np.linspace(0.05,30,250)
y = np.tanh(x)
```

```python
alt_min = 0.9
alt_max = 1.2

d = (alt_min+alt_max)/2
a = d - alt_min
c = 2.4
b = 0.5
```

Função que cria o gráfico f(v) (sensibilidade x velocidade)

```python
def f(x,a,b,c,d):
    y =  a*np.tanh(b*x-c) + d
    return y
```

Função que plota

```python
def plot(c,b,alt_min, alt_max):
    d = (alt_min+alt_max)/2
    a = d - alt_min
    global x
    y = f(x,a,b,c,d)
    #y =  a*np.tanh(b*x-c) + d
    plt.axvline(20, label = 'Velocidade extrema')
    plt.axvline(10, label = 'Movimento rápido', c = 'dodgerblue')
    plt.plot(x,y, label = 's(v)',color = 'blueviolet')
    plt.legend()
    plt.xlabel('Velocidade')
    plt.ylabel('Sensibilidade')
    plt.grid(True)
    plt.xlim(0,30)
    plt.ylim(0.8,1.3)
```

Gráfico interativo - use isso para descobrir os melhores valores para as constantes

```python
interact(plot,c=FloatSlider(value=c, min = 0.1, max = 5, step = 0.01),
         b=FloatSlider(value=b, max = 2.5, step = 0.001, readout_format='.3f'),
         alt_min=FloatSlider(value=alt_min, min = 0.8, max = 1.3, step = 0.01),
         alt_max=FloatSlider(value=alt_max, min = 1, max = 1.6, step = 0.01))   
```
